```python
import pandas as pd

raw_df = pd.read_csv("Dataset/SCDB_2021.csv")

raw_df.head()
```

```python
print(raw_df.shape)
```

```python
# Correlation Matrix
import numpy
from matplotlib import pyplot

correlations = raw_df.corr()
fig = pyplot.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
pyplot.show()
```

```python
# Pairwise Pearson correlations
from pandas import set_option

set_option('display.width', 100)
set_option('precision', 3)

correlations = raw_df.corr(method='pearson')
print(correlations)
```
